```python
import pickle
import os
import pandas as pd
from collections import defaultdict
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def load_pickle(file_path,add_data_path=True):
        if add_data_path:
            file_path=os.path.join(file_path,'data/history')
            
        
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        return data


distributions=['rnd_graph_800vertices_unweighted',
               'rnd_graph_800vertices_weighted',
               'planar_800vertices_unweighted',
               'planar_800vertices_weighted',
               'toroidal_grid_2D_800vertices_weighted',
               ]

models=['ECO_DQN','LSDQN','LinearRegression']


dataframe={'model':[],'distribution':[],'opt_step':[]}
for model in models:
    print(model)
    for distribution in distributions:
        data=load_pickle(f'pretrained agents/{distribution}_{model}')
        opt_step=0

        for i in range(len(data)):
            scores=np.array(data.iloc[i]['scores']).T
            opt_step+=np.average(np.average(np.argmax(scores,axis=1)))
            # print(np.average(np.average(np.argmax(scores,axis=1))))
        # print(opt_step/len(data)/scores.shape[1])
        mean_opt=opt_step/len(data)/scores.shape[1]
        dataframe['model'].append(model)
        dataframe['distribution'].append(distribution)
        dataframe['opt_step'].append(mean_opt)



```

```python
df=pd.DataFrame(dataframe)
```
