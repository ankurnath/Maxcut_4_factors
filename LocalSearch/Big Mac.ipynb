```python
import os
from collections import defaultdict
from glob import glob
import numpy as np
from scipy.sparse import csr_matrix,save_npz,load_npz
from experiments.utils import load_graph_set, mk_dir
```

```python
file_paths=glob('../rudy_all/*')

def parse_graph(file_path):
    adjacency_matrix = None
    with open(file_path, 'r') as file:
        lines = file.readlines()
        n, m = map(int, lines[0].split())
        adjacency_matrix = [[0] * n for _ in range(n)]
        for line in lines[1:]:
            h, t, weight = map(int, line.split())
            adjacency_matrix[h - 1][t - 1] = weight
            adjacency_matrix[t - 1][h - 1] = weight  # Assuming undirected graph
    return adjacency_matrix


cnts=defaultdict (int)


file_paths.sort()
for file_path in file_paths:
    
    filename=os.path.basename(file_path)
    adjacency_matrix = parse_graph(file_path)

    distribution=filename.split('_')[0]

    folder=f'../data/testing/BigMac_{distribution}'
    mk_dir(folder)
    save_file_name = filename
    save_file_path=os.path.join(folder,save_file_name)
    sparse_matrix = csr_matrix(adjacency_matrix)
    cnts[distribution]+=1
    save_npz(save_file_path, sparse_matrix)

    
    
    
```
